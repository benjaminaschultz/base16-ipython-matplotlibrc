```python
N=255
M=N//3
reds = np.concatenate((np.linspace(b16red[0],b16blue[0],M),np.linspace(b16blue[0],b16green[0],M)))
blues = np.concatenate((np.linspace(b16red[1],b16blue[1],M),np.linspace(b16blue[1],b16green[1],M)))
greens = np.concatenate((np.linspace(b16red[2],b16blue[2],M),np.linspace(b16blue[2],b16green[2],M)))
```

```python
colors = np.array((reds,blues,greens)).transpose()
```

```python
from matplotlib.colors import ListedColormap
cmap = ListedColormap(colors,name='b16')
from matplotlib import cm
cm.register_cmap('b16_jet',cmap)
rcParams['image.cmap']='b16_jet'
```
